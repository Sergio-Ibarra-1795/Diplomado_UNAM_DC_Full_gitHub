Importamos las librerias necesarias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture 
```

Vamos a importar la data 

```python
X_train = np.load('data.npy')
X_train.shape
```

Vamos a graficar nuestra data a agrupar 

```python
plt.scatter(X_train[:,0], X_train[:,1])
plt.axis('equal')
plt.show()
```

Creamos el modelo gmm 

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X_train)
```

Podemos obtener medias y covarianzas del modelo gmm entrenado

```python
print('Medias: \n', gmm.means_)
print('Covariances: \n', gmm.covariances_)
```

```python
### Vamos a generar una maya de puntos
X,Y = np.meshgrid(np.linspace(-1,6), np.linspace(-1,6))

### Vamos a generar una maya de puntos
XX = np.array([X.ravel(), Y.ravel()]).T
##Puntos de X e Y pegados como si fuera 1 dimensión 
Z = gmm.score_samples(XX)
Z = Z.reshape((50,50))

plt.contour(X,Y,Z)
plt.scatter(X_train[:,0], X_train[:,1])
plt.show()
```
