```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import string


# Tokenize the text
tokens = word_tokenize(moby_raw)

# Filter out punctuation tokens
punctuation_tokens = [token for token in tokens if token in string.punctuation]

# Count the frequency of each punctuation token
punctuation_freq = FreqDist(punctuation_tokens)

# Print the frequency distribution of punctuation tokens
print(punctuation_freq)
```

```python
for punctuation, frequency in punctuation_freq.items():
    print(f"{punctuation}: {frequency}")
```

The result <FreqDist with 15 samples and 34408 outcomes> means that there are 15 unique punctuation tokens in your text, and a total of 34,408 individual occurrences of these punctuation tokens.

To break down the interpretation further:

"15 unique samples": This refers to the number of distinct punctuation tokens present in the text. In other words, there are 15 different types of punctuation marks used.

"34408 outcomes": This represents the total count of punctuation tokens in the text. It's the sum of the occurrences of all the individual punctuation marks.

In this context, "samples" are the different types of punctuation marks (e.g., '.', ',', '!', '?', etc.), and "outcomes" are the total occurrences of these punctuation marks in the text.

### Pregunta 5

Encontrar la palabra mas larga en texto1 y la longitud de dicha palabra.

*Esta funcin debe devolver una tupla `(palabra, longitud)`.*

```python
def respuesta5():
    words_len = [(word, len(word)) for word in set(text1.tokens)]
    respuesta = sorted(words_len, key=lambda word_len: word_len[1], reverse=True)
    return respuesta[0]

respuesta5()
```

### Pregunta 6

Qu palabras nicas tienen una frecuencia de ms de 2000? Cul es su frecuencia?

"Sugerencia: es conveniente utilizar` isalpha () `para comprobar si el token es una palabra y no una puntuacin".

* Esta funcin debe devolver una lista de tuplas de la forma `(frecuencia, palabra)` ordenadas en orden descendente de frecuencia. *

```python
def respuesta6():
    fdist = FreqDist(text1)
    words_unique = [(fdist[word], word) for word in set(text1.tokens) if word.isalpha() and fdist[word] > 2000]
    respuesta = sorted(words_unique, key=lambda word: word[0], reverse=True)
    return respuesta

respuesta6()
```

### Pregunta 7
Cul es el nmero promedio de tokens por oracin?

```python
import statistics
def respuesta7():
    sentences = nltk.sent_tokenize(moby_raw)
    respuesta = [len(nltk.word_tokenize(w)) for w in sentences]
    return statistics.mean(respuesta)

respuesta7()
```
