## C)     ¿Qué cantidad de sustantivos, adjetivos y verbos posee el texto?

### Primero con NLTK

Un intento de hacer con la implementación en inglés

```python
import nltk
nltk.download('averaged_perceptron_tagger')  # Download the POS tagging data
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import Counter

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


# Perform POS tagging on the words
pos_tags = pos_tag(tokens_ramo_raw_words)

# Create an empty Counter for each part of speech
noun_counter = Counter()
adjective_counter = Counter()
verb_counter = Counter()

# Loop through the tagged words and count nouns, adjectives, and verbs
for word, tag in pos_tags:
    if tag.startswith('NN'):  # Noun
        noun_counter.update([word])
    elif tag.startswith('JJ'):  # Adjective
        adjective_counter.update([word])
    elif tag.startswith('VB'):  # Verb
        verb_counter.update([word])

# Print the counts for each part of speech
print(f"Nouns: {noun_counter}")
print(f"Adjectives: {adjective_counter}")
print(f"Verbs: {verb_counter}")
```

```python
# Calculate the total counts
total_nouns = sum(noun_counter.values())
total_adjectives = sum(adjective_counter.values())
total_verbs = sum(verb_counter.values())

# Print the total counts for each part of speech
print(f"Total Nouns: {total_nouns}")
print(f"Total Adjectives: {total_adjectives}")
print(f"Total Verbs: {total_verbs}")
```

### Ahora con el moduo SPACY

```python
import spacy
from collections import Counter

# Load the Spanish language model in SpaCy
nlp = spacy.load('es_core_news_sm')


# Join the tokens to form a string for SpaCy processing
text = ' '.join(tokens_ramo_raw_words)

# Process the text using SpaCy
doc = nlp(text)

# Create an empty Counter for each part of speech
noun_counter = Counter()
adjective_counter = Counter()
verb_counter = Counter()

# Loop through the processed tokens and count nouns, adjectives, and verbs
for token in doc:
    if token.pos_ == 'NOUN':  # Noun
        noun_counter.update([token.text])
    elif token.pos_ == 'ADJ':  # Adjective
        adjective_counter.update([token.text])
    elif token.pos_ == 'VERB':  # Verb
        verb_counter.update([token.text])

# Print the counts for each part of speech
print(f"Nouns: {noun_counter}")
print(f"Adjectives: {adjective_counter}")
print(f"Verbs: {verb_counter}")
```

```python
# Calculate the total counts
total_nouns = sum(noun_counter.values())
total_adjectives = sum(adjective_counter.values())
total_verbs = sum(verb_counter.values())

# Print the total counts for each part of speech
print(f"Total Nouns: {total_nouns}")
print(f"Total Adjectives: {total_adjectives}")
print(f"Total Verbs: {total_verbs}")
```
