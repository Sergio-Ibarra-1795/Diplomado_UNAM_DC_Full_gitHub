## A)    ¿Cuántas palabras hay en el texto?

```python
# Usaremos expresiones regulares para eliminar signos de puntuación y contar las palabras
import re

libro_limpio = re.sub(r'[^\w\s]','',libro)
print(libro_limpio[0:1500])
```

```python
# Tokenizar un texto, produce una lista de palabras y puntuaciones
tokens = nltk.word_tokenize(libro_limpio, language='spanish')
print( type(tokens), len(tokens) )
print( tokens )
```

### Se tienen 826 palabras en el texto

## B) ¿Cuántas palabras diferentes existen?

```python
# Eliminemos los duplicado en tokens
palabras_unicas = list(set(tokens))
print(palabras_unicas)
```

```python
len(palabras_unicas)
```

### Se tienen 466 palabras diferentes

## C) ¿Qué cantidad de sustantivos, adjetivos y verbos posee el texto?

#### C.1. Usando metodo NLTK

```python
# Aplicamos la función POS a la lista de tokens
tokens_pos = nltk.pos_tag(tokens)
```

```python
# Revisemos los tipos de tokens encontrados
for i in range(0,len(tokens_pos)):
  print(tokens_pos[i])
```

```python
# Contando cuantos sustantivos, adjetivos y verbos se tienen
sustantivos = 0
adjetivos = 0
verbos = 0
for i in range(0,len(tokens_pos)):
  if (tokens_pos[i][1]=='NNP'):
    sustantivos += 1
  if (tokens_pos[i][1]=='NN'):
    adjetivos += 1
  if (tokens_pos[i][1]=='VBZ'):
    verbos += 1

print('El número de sustantivos es: ', sustantivos)
print('El número de adjetivos es: ', adjetivos)
print('El número de verbos es: ', verbos)
```

```python
from spacy.lang.es.examples import sentences

nlp = spacy.load('es_core_news_sm')
doc = nlp(libro)
print(doc.text)
tipos_token = []
for token in doc:
    print(token.text, token.pos_, token.dep_)
    tipos_token.append(token.pos_)
```

```python
print(tipos_token)
```

```python
print('El número de sustantivos es: ', tipos_token.count('NOUN'))
print('El número de adjetivos es: ', tipos_token.count('ADJ'))
print('El número de verbos es: ', tipos_token.count('VERB'))
```

#### C.3. Usando metodo FREELING

```python
# Haciendo el mismo ejercicio con Freeling
import requests
import pandas as pd

# Archivo a ser enviado
files = {'file': open('ElRamoAzul.txt', 'rb')}

# Parámetros
params = {'outf': 'tagged', 'format': 'json'}
# params = {'outf': 'dep', 'format': 'json'}

# Enviar petición
url = "http://www.corpus.unam.mx/servicio-freeling/analyze.php"
r = requests.post(url, files=files, params=params)
lista_tags_freeling = []   ## Aquí guardaremos los tipos de token (tags)
try:
    # Convertir de formato json
    obj = r.json()

    # Ejemplo, obtener todos los lemas
    for sentence in obj:
        for word in sentence:
            # print (word)
            print (word['lemma'],word['token'],word['tag'])
            lista_tags_freeling.append(word['tag'])
except:
    import traceback
    print( traceback.format_exc() )
```

```python
# Lista de tags obtenidos
lista_tags_freeling
```

```python
# Contador de elementos que empiezan con "NC"
contador_NC = 0
for elemento in lista_tags_freeling:
    if elemento.startswith("NC"):
        contador_NC += 1
```

```python
# Contador de elementos que empiezan con "AQ"
contador_ADJ = 0
for elemento in lista_tags_freeling:
    if elemento.startswith("AQ"):
        contador_ADJ += 1
```

```python
# Contador de elementos que empiezan con "VM"
contador_VRB = 0
for elemento in lista_tags_freeling:
    if elemento.startswith("VM"):
        contador_VRB += 1
```

```python
# Contamos la cantidad de sustantivos, adjetivos y verbos
print('El número de sustantivos es: ', contador_NC)
print('El número de adjetivos es: ', contador_ADJ)
print('El número de verbos es: ', contador_VRB)
```

#### Tabla comparativa

```python
# Construimos la tabla para comparar los métodos
import pandas as pd

tabla = pd.DataFrame()

tabla['NLTK'] = None
tabla['Spacy'] = None
tabla['Freeling'] = None

tabla.loc['Sustantivos'] = [sustantivos,tipos_token.count('NOUN'),contador_NC]
tabla.loc['Adjetivos'] = [adjetivos,tipos_token.count('ADJ'),contador_ADJ]
tabla.loc['Verbos'] = [verbos,tipos_token.count('VERB'),contador_VRB]

tabla
```

### De la tabla comparativa observamos que NLTK es el qe menos verbos encuentra, observando los datos notamos que en efecto algunos verbos los clasifica como sustantivos. Spacy y Freeling clasificaciones en el mismo orden, más sustantivos, seguidos de verbos y mucho menos adjetivos. La ventaja de Freeling es que considera diferentes opciones para cada categoría.
