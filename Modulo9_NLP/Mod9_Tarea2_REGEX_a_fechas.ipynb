Se construye la función que extrae todas las fechas 

```python
import re

def extract_dates(filename):
  dates = []

  # Match more date patterns 
  date_regex = re.compile(r'(((\d{1,2} [A-Za-z]{3,9} \d{2,4})|(\d{1,2}/\d{1,2}/\d{2,4})|(\d{1,2}/\d{4})|(\d{1,2}|\w{3,9}) (\d{4})|(\w{3,9} \d{2,4})))')

  with open(filename) as f:
    for line in f:
      found_dates = date_regex.findall(line)  
      for date in found_dates:
        dates.append(date[0])

  return dates
```

De los elementos en dates, hay algunos que no son fechas, limpiemos a nuestra lista de todos esos elementos 

```python
import re

dates = dates # list of dates

# Words to filter out
filter_words = [r'\bdelivery\b', r'\bphrectomy\b', r'\bfor\b', r'\bB12\b', r'\bAge\b', r'\bage\b', r'\bsince\b', r'\battempt\b', r'\bcall\b', r'\bDisorders\b', r'\bTrazodone\b', r'\bDrinking\b', r'\bEly\b', r'\bZoloft\b', r'\bfrom\b', r'\blost\b', r'\bweighed\b', r'\bWoriking\b', r'\bweighed\b', r'\bpleasant\b', r'\bsleeve\b', r'\bstarted\b', r'\bresection\b', r'\bplaced\b', r'\budication\b', r'\brednisone\b', r'\bedication\b', r'\bdone\b', r'\bSince\b', r'\bquit\b', r'\boxetine\b', r'\bssee\b', r'\bluoxetine\b', r'\bbupropion\b', r'\bstarting\b', r'\bsee\b', r'\bAbilify\b', r'\bertraline\b', r'\beby\b', r'\bbupropion\b', r'\bssee\b', r'\bwas\b', r'\btrazadone\b', r'\bellbutrin\b', r'\bSeroquel\b', r'\bZoloft\b', r'\bnlafaxine\b', r'\bEly\b', r'\bZolof\b', r'\bweighed\b', r'\btwins\b', r'\bsleeve\b', r'\bpleasant\b', r'\budication\b', r'\brednisone\b', r'\bQTc\b', r'\bedication\b', r'\bSince\b', r'\bsSince\b',    r'\bdON\b',  r'\bafter\b',  r'\bpOct\b',  r'\bover\b',  r'\bProzac\b',  r'\buntil',  r'\bcalling\b', r'\bgerd\b',  r'\bLIVEF\b',  r'\bdose\b', r'\blast\b', r'\bages\b']

# Build regex
filter_regex = re.compile('|'.join(filter_words))


# Filter the list 
cleaned_dates = [d for d in dates if not filter_regex.search(d)]
```

Vamos a aplicar las reglas: 
Todas las fechas están en formato xx/xx/xx son mm/dd/aa
Todas las fechas en las que el año está codificado en solo dos dígitos corresponden a años posteriores a la década de 1900 (p. Ej., 1/5/89 es el 5 de enero de 1989).
Si falta el día (p. Ej., 9/2009), suponga que es el primer día del mes (p. Ej., septiembre, 1 de 2009).
Si falta el mes (por ejemplo, 2010), suponga que es el primero de enero de ese año (p. Ej., enero, 1 de 2010).
Tenga cuidado con los posibles errores tipográficos, ya que este es un conjunto de datos derivados de la vida real.
Con el fin de tener todas las fechas en el formato mm/dd/aa

```python
import re
import calendar


cleaned_dates = cleaned_dates


corrected_dates = []

for date in cleaned_dates:

  # Handle full month name
  month_pattern = r'([a-zA-Z]+) (\d{1,2}),? (\d{2,4})'

  def get_month(x):
    try:
      month = list(calendar.month_abbr).index(x.group(1)[:3].title()) 
    except ValueError:
      month = 1
    
    return x.group(2) + '/' + str(month) + '/' + x.group(3)

  date = re.sub(month_pattern, lambda x: get_month(x), date)

  # Handle month and day  
  month_day_pattern = r'([a-zA-Z]+) (\d{1,2})'

  def get_month_day(x):

    try:
      month = list(calendar.month_abbr).index(x.group(1)[:3].title())
    except ValueError:
      month = 1

    if len(x.groups()) > 2:
      return x.group(2) + '/' + str(month) + '/' + x.group(3)
    else:
      return x.group(2) + '/' + str(month) + '/2000'

  date = re.sub(month_day_pattern, lambda x: get_month_day(x), date)

  # Rest of logic

  # Handle missing day
  if re.match(r'^\d{1,2}/\d{4}$', date):
    date = '1/' + date


  # Handle missing month
  elif re.match(r'^\d{4}$', date):
    date = '1/1/' + date 


  # Handle missing year
  elif re.match(r'^\w+ \d{1,2}$', date):
    date = date + '/2000'

  
  # Reformat 2-digit year only
  date = re.sub(r'/(\d{2})$', '/\\1', date)

  # Pad 2-digit years
  if re.search(r'/\d{2}$', date): 
    year = int(date[-2:])
    if year <= 99 and year >= 30:
      date = re.sub(r'/(\d{2})$', '/19\\1', date)
    else:
      date = re.sub(r'/(\d{2})$', '/20\\1', date)

  
  # Flip month/day
  date = re.sub(r'^(\d{1,2})/(\d{1,2})', '\\2/\\1', date)


  # Add leading zeros 
  date = re.sub(r'^(\d{1})/', '0\\1/', date)
  date = re.sub(r'^(\d{1})/(\d{1})/', '0\\1/0\\2/', date)

  
  # Convert 4-digit year
  if len(date) > 6:
    date = re.sub(r'/(\d{4})$', '/\\1', date)


  corrected_dates.append(date) 

print(corrected_dates)
```
