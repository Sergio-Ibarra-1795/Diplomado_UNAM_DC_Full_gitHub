### Usando NLTK

```python
import nltk
import traceback
import re

#Archivo a ser enviado
with open('ElRamoAzul.txt', 'r', encoding="utf-8") as f:
    text = f.read()

try:
    tag = nltk.pos_tag( nltk.word_tokenize(text) )
    
    #inicializacion de arreglo
    array_words = []

    #inicializacion de variables contadoras
    num_noun = 0#etiqueta categoria sustantivo NN
    num_adj = 0 #etiqueta categoria adjetivo JJ
    num_verb = 0 #etiqueta categoria verbo VB
    pattern = re.compile('(?:[.,\/!$%?¿?!¡\^&\*;:{}=><\-\–_`~()”“"\'\|])')
    
    for t in tag:
        token = t[0]
        #no se tienen en cuanta los simbolos de puntacion para contarlas como palabras
        if pattern.search(token) is None:
            #cuantas palabras hay.
            array_words.append(token)
            #Contar el número de sustantivos, adjetivos y verbos.
            if t[1].startswith('NN'):
                num_noun +=1
            elif t[1].startswith('JJ'):
                num_adj +=1
            elif t[1].startswith('VB'):
                num_verb +=1

    print("En el texto <<El ramo azul>>, de Octavio Paz")
    print("- Existen "+str(len(array_words))+" palabras")
    print("- De ellas, "+str(len(set([w.lower() for w in array_words])))+" son palabras diferentes")
    print("- En total hay "+str(num_noun)+" sustantivos")
    print("- En total hay "+str(num_adj)+" adjetivos")
    print("- En total hay "+str(num_verb)+" verbos")

except:
    print( traceback.format_exc() )
```

```python
' - '.join(array_words)
```

### Usando sólo el Servicio de Freeling

```python
import requests
import traceback
import os
import re

#Archivo a ser enviado
files = {'file': open('ElRamoAzul.txt', 'rb')}

#Parámetros
params = {'outf': 'tagged', 'format': 'json'}

#Enviar petición
url = "http://www.corpus.unam.mx/servicio-freeling/analyze.php"
r = requests.post(url, files=files, params=params)

try:
    #Convertir de formato json
    obj = r.json()

    #inicializacion de arreglo
    array_words = []

    #inicializacion de variables contadoras
    num_noun = 0#etiqueta categoria sustantivo N
    num_adj = 0 #etiqueta categoria adjetivo A
    num_verb = 0 #etiqueta categoria verbo V
    pattern = re.compile('(?:[.,\/!$%?¿?!¡\^&\*;:{}=><\-\–_`~()”“"\'\|])')

    for sentence in obj:
        for word in sentence:
            token = word["token"]
            #no se tienen en cuanta los simbolos de puntacion para contarlas como palabras
            if pattern.search(token) is None:
                #cuantas palabras hay.
                array_words.append(token)
                #Contar el número de sustantivos, adjetivos y verbos.
                if word["tag"][0] == "N":
                    num_noun +=1
                elif word["tag"][0] == "A":
                    num_adj +=1
                elif word["tag"][0] == "V":
                    num_verb +=1

    print("En el texto <<El ramo azul>>, de Octavio Paz")
    print("- Existen "+str(len(array_words))+" palabras")
    print("- De ellas, "+str(len(set([w.lower() for w in array_words])))+" son palabras diferentes")
    print("- En total hay "+str(num_noun)+" sustantivos")
    print("- En total hay "+str(num_adj)+" adjetivos")
    print("- En total hay "+str(num_verb)+" verbos")

except:
    print( traceback.format_exc() )
```

### Usando SPACY

```python
import spacy
nlp = spacy.load('es_core_news_sm')
import traceback
import re

#Archivo a ser enviado
with open('ElRamoAzul.txt', 'r', encoding="utf-8") as f:
    text = f.read()

try:
    doc = nlp(text)
    # print(doc.text)

    #inicializacion de arreglo
    array_words = []
    tw = []

    #inicializacion de variables contadoras
    num_noun = 0#etiqueta categoria sustantivo NN
    num_adj = 0 #etiqueta categoria adjetivo JJ
    num_verb = 0 #etiqueta categoria verbo VB

    for token in doc:
        #no se tienen en cuanta los simbolos de puntacion para contarlas como palabras
        if token.pos_ not in ["PUNCT","SPACE","PROPN"]:
            #cuantas palabras hay.
            array_words.append(token.text)
            tw.append([token.text, token.pos_])
            #Contar el número de sustantivos, adjetivos y verbos.
            if token.pos_ == "NOUN":
                num_noun +=1
            elif token.pos_ == "ADJ":
                num_adj +=1
            elif token.pos_ == "VERB":
                num_verb +=1

    print("En el texto <<El ramo azul>>, de Octavio Paz")
    print("- Existen "+str(len(array_words))+" palabras")
    print("- De ellas, "+str(len(set([w.lower() for w in array_words])))+" son palabras diferentes")
    print("- En total hay "+str(num_noun)+" sustantivos")
    print("- En total hay "+str(num_adj)+" adjetivos")
    print("- En total hay "+str(num_verb)+" verbos")
        
except:
    print( traceback.format_exc() )
```
