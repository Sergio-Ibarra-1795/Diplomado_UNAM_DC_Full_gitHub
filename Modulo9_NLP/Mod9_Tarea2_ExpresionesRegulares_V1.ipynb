#1. Extracción de fechas

```python
# Importando el modulo de regex de Python
import re
from datetime import datetime
```

```python
# Abrimos archivo de fechas
with open('dates.txt', 'r', encoding='utf-8') as f:
    notas_contents = f.read()
```

```python
# Separamos cada linea
notas = notas_contents.split('\n')
```

```python
# Definiendo función para extraer fecha de cada nota
def extraeFecha(nota):
  fecha = []

  #Patron 3 - OK
  #20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
  patron = re.compile(r'\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[,.]* \d{4}')
  fecha += patron.findall(nota)

  #Patron 2 - OK
  #Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;
  patron = re.compile(r'(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[.-]* \d{1,2}(?:,)? \d{4}')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  #Patron 4 - OK
  #Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
  patron = re.compile(r'(?:Ene|Feb|Mar|Abr|May|Jun|Jul|Ago|Sep|Oct|Nov|Dic ) \d{1,2}(?:nd, |th, |st, )?\d{4}')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  #Patron 5 - OK
  #Feb 2009; Sep 2009; Oct 2010
  patron = re.compile(r'(?:Ene|Feb|Mar|Abr|May|Jun|Jul|Ago|Sep|Oct|Nov|Dic ) \d{4}')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  #Patron 1 - OK
  #04/20/2009; 04/20/09; 4/20/09; 4/3/09
  patron = re.compile(r'(?:\d{1,2}[-/]\d{1,2}[-/])(?:19\d{2}|20\d{2}|[5-9]\d{1})')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  #Patron 6 - OK
  #6/2008; 12/2009
  patron = re.compile(r'\d{1,2}/\d{4}')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  #Patron 7 - OK
  #2009; 2010
  patron = re.compile(r'19\d{2}|20\d{2}')
  fecha += patron.findall(nota) if len(fecha) == 0 else []

  return fecha[0] if fecha != [] else ''
```

```python
# Genera lista con fechas
#fechas = [ max(fechas) for fechas in [ extraeFecha(nota) for nota in notas ] if len(fechas) != 0 ]
fechas = [ extraeFecha(nota) for nota in notas ]
fechas = [ fecha for fecha in fechas if fecha != '']
```

#2. Reglas de negocio

```python
#Se normalizan las fechas de todos los formatos a solo un formato Primero el año para poder ordenarlo por año, mes y día
def normalizar_fechas(date_str):
    formatos = [
        "%Y-%m-%d", "%d/%m/%Y", "%d-%m-%Y", "%m-%d-%Y", "%m/%d/%y", "%m-%d-%y", "%m/%d/%Y", "%Y/%m/%d", "%d %B %Y",
        "%m %Y", "%m/%Y", "%m/%y", "%Y", "%d %b %Y", "%b %d, %Y", "%B %d, %Y", "%B. %d %Y", "%b %d %Y", "%B. %d, %Y",
        "%B %d %Y", "%b %d %Y", "%b %Y", "%B %Y", "%b. %d, %Y", "%d de %B de %Y" ]

    for fmt in formatos:
        try:
            date_obj = datetime.strptime(date_str,fmt)
            normalized_date = date_obj.strftime("%Y/%m/%d")
            return normalized_date
        except ValueError:
            pass
    return ''

normalized_dates = [ normalizar_fechas(fecha) for fecha in fechas]
print(normalized_dates)
```

```python
#Ordenar Las fechas en orden cronologico
sorted_dates = sorted(normalized_dates)

#Cambiando a formato requerido ("%d/%m/%y") y ordenado
fechasFinales = [ datetime.strptime(fecha,"%Y/%m/%d").strftime("%d/%m/%y") for fecha in sorted_dates]
print(fechasFinales)
```
