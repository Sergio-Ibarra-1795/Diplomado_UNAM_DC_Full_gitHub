```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer

tabla_breast_cancer = load_breast_cancer(as_frame=True)
```

```python
tabla_breast_cancer.frame
```

Separamos en variable dependiente (y) y variables independientes (X)

```python
y_breast_cancer = tabla_breast_cancer['target']
y_breast_cancer
```

```python
tabla_breast_cancer['frame'].columns
```

```python
X_breast_cancer = tabla_breast_cancer['data'][['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']]
X_breast_cancer
```

Se separa en train y test la data de breast cancer 

```python
## Separacion en train y test 
from sklearn.model_selection import train_test_split

X_breast_cancer_train, X_breast_cancer_test, y_breast_cancer_train, y_breast_cancer_test = train_test_split(X_breast_cancer,y_breast_cancer,test_size=0.2,random_state = 2022)
X_breast_cancer_test
```

Importamos y usamos el modelo de LogisticRegression 

```python
from sklearn.linear_model import LogisticRegression

modelo_Logistic_Regression = LogisticRegression()
modelo_Logistic_Regression_breast_cancer = modelo_Logistic_Regression.fit(X_breast_cancer_train, y_breast_cancer_train)
modelo_Logistic_Regression_breast_cancer
```

Hacemos uso del modelo_Logistic_Regression_breast_cancer

```python
ypred_modelo_Logistic_Regression_breast_cancer = modelo_Logistic_Regression.predict(X_breast_cancer_test)
ypred_modelo_Logistic_Regression_breast_cancer
```

Evaluamos el modelo_Logistic_Regression_breast_cancer mediante confusion_matrix, accuracy y presicion 

##### Matriz confusión para modelo_Logistic_Regression_breast_cancer

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_breast_cancer_test,ypred_modelo_Logistic_Regression_breast_cancer)
```

##### Accuracy para modelo_Logistic_Regression_breast_cancer

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_breast_cancer_test,ypred_modelo_Logistic_Regression_breast_cancer)
```

##### Precision para modelo_Logistic_Regression_breast_cancer

```python
from sklearn.metrics import precision_score

precision_score(y_breast_cancer_test,ypred_modelo_Logistic_Regression_breast_cancer)
```

Se observa que el modelo LogisticRegression aplicado a datos de breast cancer arrojar una exactitud y precisión de clasificación de ~90% 

##### Gráficamos el resultado de la Regression Logistica para la data de breast cancer 

```python
import matplotlib.pyplot as plt
import numpy as np

# Select two features for visualization
feature1 = 'mean radius'
feature2 = 'mean texture'

# Extract the corresponding columns from the training data
x_points = X_breast_cancer_train[[feature1, feature2]]
y_points = y_breast_cancer_train

# Train a new logistic regression model using only two features
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_points, y_points)

# Create a meshgrid to generate the decision boundary
x_min, x_max = x_points[feature1].min() - 1, x_points[feature1].max() + 1
y_min, y_max = x_points[feature2].min() - 1, x_points[feature2].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Make predictions on the grid points
decision_vals = model.predict(grid_points)

# Reshape decision_vals to match the meshgrid shape
decision_vals = decision_vals.reshape(xx.shape)

# Plot the decision boundary and scatter points
plt.figure()
plt.contourf(xx, yy, decision_vals, alpha=0.8)
plt.scatter(x_points[feature1], x_points[feature2], c=y_points, edgecolors='k')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title('Decision Boundary')
plt.show()
```
