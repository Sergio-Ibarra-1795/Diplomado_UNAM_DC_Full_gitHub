Vamos a importar la base de datos 

```python
import pandas as pd

url = "https://gitlab.com/dgtic5/res/-/raw/main/aprendizajeSupervizado/mushrooms.csv"

tabla_hongos = pd.read_csv(url)
tabla_hongos
```

Y se pueden observar los likelihood agrupando por clase 

```python
tabla_hongos.groupby(["class","cap-shape"]).size()
```

Vamos a aplicar OneHotEncoding 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y = tabla_hongos["class"]
X = tabla_hongos.drop(["class"],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 2023)
X_train
```

Se define el codificador y se entrena 

```python
codificador = OneHotEncoder()
codificador.fit(X_train)
```

Aplicamos el codificador 

```python
#codificador.fit(X_train)
#X_train = codificador.transform(X_train)
X_train = codificador.fit_transform(X_train)

X_test = codificador.transform(X_test)
```

```python
X_train.toarray()
```

Ahora si con los datos ya encodificados hacemos el modelo Naive Bayes 

```python
from sklearn.naive_bayes import MultinomialNB

modelo_NB = MultinomialNB(alpha = 0)
modelo_NB.fit(X_train,y_train)

y_pred_modelo_NB = modelo_NB.predict(X_test)
y_pred_modelo_NB
```

Evaluemos el primer modelo NB de clasificaicón

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_modelo_NB))
```

Veamos el comportamiento con un valor diferente de alpha para el suevizamiento 

```python
from sklearn.naive_bayes import MultinomialNB

modelo_NB_1 = MultinomialNB(alpha = 1)
modelo_NB_1.fit(X_train,y_train)

y_pred_modelo_NB_1 = modelo_NB_1.predict(X_test)
y_pred_modelo_NB_1
```

Evaluemos el segundo modelo NB de clasificaicón

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_modelo_NB_1))
```

Y se puede obtener la probabilidad de pertenencia a cada clase 

```python
modelo_NB.predict_proba(X_test)
```
