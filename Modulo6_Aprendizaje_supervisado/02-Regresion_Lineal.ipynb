### Aplicación del modelo lineal en Python

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

Se entrena al modelo con model.fit(X_train,y_train)

```python
modelo.fit(X_train,y_train)
```

```python
modelo_linea_entrenado1 = modelo.fit(X_train,y_train)
```

Como saber las pendientes o coficientes que enctró mi modelo 

```python
caracteristicas = X.columns
coef_dict = {}

for coef,caracteristica in zip(modelo.coef_,caracteristicas):
  coef_dict[caracteristica] = coef

print(coef_dict)
```

```python
modelo_linea_entrenado1.coef_
```

Ahora vamos a predecir usando el modelo de LinearRegression

```python
y_pred_modelo1 =modelo_linea_entrenado1.predict(X_test)
y_pred_modelo1
```

```python
from sklearn.metrics import mean_absolute_error

# Assuming you have y_test and y_pred_modelo1 as the true labels and predicted labels, respectively
mae_modelo1 = mean_absolute_error(y_test, y_pred_modelo1)

print("Mean Absolute Error:", mae_modelo1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_pred_train = modelo_linea_entrenado1.predict(X_train)
y_pred_train
```

```python
x_lin = np.linspace(0.5,1.1,20)
y_lin = x_lin

x_model = y_train
y_model = y_pred_train

fig, ax = plt.subplots()
ax.plot(x_lin,y_lin)
ax.scatter(x_model,y_model,s=60)
```

```python
import matplotlib.pyplot as plt

# Assuming you have already fitted a linear regression model called "modelo_linea_entrenado1"
y_pred_modelo1 # Get the predicted values

# Plot predicted values versus true values
plt.scatter(y_test, y_pred_modelo1)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--')  # Add a diagonal line for reference
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Predicted Values versus True Values')
plt.show()

# Plot residuals
residuals = y_test - y_pred_modelo1# Calculate residuals
plt.scatter(y_pred_modelo1, residuals)
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals')
plt.show()
```
