```python
# Convert 'company_size' column to categorical data type
ds_salaries['experience_level'] = pd.Categorical(ds_salaries['experience_level'])

# Count the occurrences of each category
category_counts = ds_salaries['experience_level'].value_counts()

# Set the figure size
plt.figure(figsize=(10, 5))

# Calculate the histogram counts and bin edges using np.histogram
counts, bins = np.histogram(ds_salaries['experience_level'].cat.codes, bins=len(category_counts))

# Plot the histogram using plt.bar
plt.bar(range(len(category_counts)), counts, edgecolor='black', align='center')

# Add the count labels on top of each bar
for i, v in enumerate(counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=12)

# Set the x-axis tick positions and labels
plt.xticks(range(len(category_counts)), category_counts.index)

# Rotate the x-axis labels
plt.xticks(rotation=90)

plt.xlabel('experience_level')
plt.ylabel('Count')
plt.title('Distribution of experience_level')

plt.tight_layout()
plt.show()
```

```python
variables_numericas =   [i for i in ds_salaries.columns if ds_salaries[i].dtype=='int64' ]
variables_numericas 
```

Se calculas los outliers para 'salary'

```python
q1_salary = ds_salaries['salary'].quantile(0.25)
q1_salary

q3_salary = ds_salaries['salary'].quantile(0.75)
q3_salary

IQR_salary = q3_salary - q1_salary
IQR_salary


### Se calculan los limites inferior y superior como q1/q3 -/+ 1.5*IQR 

limite_inferior_salary = q1_salary - 1.5*IQR_salary

limite_superior_salary = q3_salary + 1.5*IQR_salary

limite_superior_salary

```

```python
ds_salaries['salary'].describe()
```

```python
salary_outliers = [salary for salary in ds_salaries['salary'] if salary > limite_superior_salary]
salary_outliers
```
