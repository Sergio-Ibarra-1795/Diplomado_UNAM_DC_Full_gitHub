##### Se calculas los outliers para 'salary'

Primero damos un vistazo a los estadísticos más clásicos de todo el array de nombre ds_salaries['salary']

```python
# Set the display format for float values
pd.options.display.float_format = '{:,.0f}'.format


ds_salaries['salary'].describe()
```

Calculamos los limites superior en infenrior con base en el Rango Interquartilico 

```python
q1_salary = ds_salaries['salary'].quantile(0.25)
q1_salary

q3_salary = ds_salaries['salary'].quantile(0.75)
q3_salary

IQR_salary = q3_salary - q1_salary
IQR_salary


### Se calculan los limites inferior y superior como q1/q3 -/+ 1.5*IQR 

limite_inferior_salary = q1_salary - 1.5*IQR_salary

limite_superior_salary = q3_salary + 1.5*IQR_salary

limite_superior_salary

```

Se observa el el valor de limite_superior_salary es de 300,000. Por lo tanto, todos los salarios superiores a ese valor, serían considerados outliers bajo la ecuación  q3_salary + 1.5*IQR_salary.
En este caso no se toma en cuenta el limite_inferior_salary, pues debido a los valores de q1_salary e IQR_salary, el valor de limite_inferior_salary es menor que cero y eso no tiene sentido en el contexto de sueldos y salarios. 


Se van a separar solo aquellos elementos de salary que tengan un valor mayor a limite_superior_salary

```python
salary_outliers = [salary for salary in ds_salaries['salary'] if salary > limite_superior_salary]
```

```python
# Calculate descriptive statistics using numpy
count = len(salary_outliers)
mean = np.mean(salary_outliers)
std = np.std(salary_outliers)
minimum = np.min(salary_outliers)
maximum = np.max(salary_outliers)
percentiles = np.percentile(salary_outliers, [25, 50, 75])

# Print the descriptive statistics
print("Count:", count)
print("Mean:", mean)
print("Standard Deviation:", std)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("25th Percentile:", percentiles[0])
print("50th Percentile (Median):", percentiles[1])
print("75th Percentile:", percentiles[2])
```

Se observa como solo 113 de 3,750 elementos tiene un valor > limite_superior_salary.
Además es evidente como la media, max y min  de este grupo de salarios es mucho más grande que la media, max y min del total de salarios.
En este caso los 113 elementos NO LOS CONSIDERARÍA OUTLIERS debido a que salary se encuentra en la moneda local de los paises, y la gran mayoria de los sueldos son en dolares, los paises que tengan una moneda "muy debil" tenderán a tener "salarios muy inflados" en la moneda local. (Ejemplo desafortunado el caso de nuestros amigos Venezolanos o Argentinos que seguro tiene sueldos en millones en su moneda local). Esta idea se refuerza con la gráfica (Histogram of salary distribution) en la que claramente se observan muchos valores en el rango de 100,000-300,000 y unos pocos en el rango mayor a 300,000 hasta millones  de unidades monetarias locales.
