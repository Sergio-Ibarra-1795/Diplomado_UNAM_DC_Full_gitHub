Hay un total de 11 variables

- remote_ratio: Se refiere al % del tiempo o % de dedicación al trabajo que se espera el trabajo haga de manera remota 
- company_location: Se refiere al lugar en donde se localiza la compañia 
- company_size: Se refiere al número de empleados que tiene la compañia
