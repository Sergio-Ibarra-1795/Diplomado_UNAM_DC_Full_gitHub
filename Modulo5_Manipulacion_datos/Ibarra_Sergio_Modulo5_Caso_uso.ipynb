3.- Gráfica la frecuencia de 3 variables categóricas que elijas, comenta la razón para haberlas elegido.

```python
# Count the occurrences of each category
category_counts = ds_salaries['experience_level'].value_counts()

# Plot the histogram
plt.hist(ds_salaries['experience_level'], bins=len(category_counts), edgecolor='black', density=False)

# Add the count labels on top of each bar
for i, v in enumerate(category_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Rotate the x-axis labels
plt.xticks(rotation=90)

plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.title('Distribution of Experience Levels')

plt.show()
```

```python
# Count the occurrences of each category
category_counts = ds_salaries['company_size'].value_counts()

# Plot the histogram
plt.hist(ds_salaries['company_size'], bins=len(category_counts), edgecolor='black', density=False)

# Add the count labels on top of each bar
for i, v in enumerate(category_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Rotate the x-axis labels
plt.xticks(rotation=90)

plt.xlabel('company_size')
plt.ylabel('Count')
plt.title('Distribution of company_size')

plt.show()
```
