Vamos a importar el archivo 

```python
import pandas as pd 
```

```python
df_clase2_csv = pd.read_csv("df_clase2.csv", encoding="utf-8")
df_clase2_csv.head()
```

Para limpiar el signo de pesos 

```python
df_clase2_csv['B1'] =df_clase2_csv['B1'].str.replace('$',"").str.replace(",","").astype(float)
```

```python
df_clase2_csv.head()
```

```python
df_clase2_csv['B2'] =df_clase2_csv['B2'].str.replace('$',"").str.replace(",","").astype(float)
```

```python
df_clase2_csv['C1'] =df_clase2_csv['C1'].str.replace('$',"").str.replace(",","").astype(float)
df_clase2_csv['C2'] =df_clase2_csv['C2'].str.replace('$',"").str.replace(",","").astype(float)
```

```python
df_clase2_csv[' C3'] =df_clase2_csv[' C3'].str.replace('$',"").str.replace(",","").astype(float)
df_clase2_csv['C4'] =df_clase2_csv['C4'].str.replace('$',"").str.replace(",","").astype(float)
```

```python
df_clase2_csv.head()
```

Para hacer sumas en las columnas 

```python
df_clase2_csv['Suma_C'] = df_clase2_csv['C1'] + df_clase2_csv['C2'] + df_clase2_csv[' C3'] + df_clase2_csv['C4']  
df_clase2_csv.head()
```

```python
df_clase2_csv['B1_calculada'] = df_clase2_csv['Suma_C'] - df_clase2_csv['B2']  
df_clase2_csv.head()
```
