## Predicción del rating usando el title

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Split data into features (titles) and target (ratings)
X_title = review_hoteles['title']
y_rating = review_hoteles['rating']
```

```python
X_title
```

Tokenizamos los valores de entrada a la red

```python
# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_title)
X_title_sequences = tokenizer.texts_to_sequences(X_title)
X_title_sequences
```

Llevamos todos los vectore de input al mismo tamanio que el vector de mayor tamanio 

```python
# Pad sequences
max_sequence_length = max([len(seq) for seq in X_title_sequences])
X_title_padded = pad_sequences(X_title_sequences, maxlen=max_sequence_length, padding='post')
X_title_padded
```

```python
X_title_padded.shape
```

Vamos a separar en train y test 

```python
from sklearn.model_selection import train_test_split

X_title_padded_train, X_title_padded_test, y_ratin_train, y_ratin_test = train_test_split(X_title_padded,y_rating,test_size=0.2,random_state=2023)
X_title_padded_train
```

```python
X_title_padded_train.shape
```

Construccion de la RNN para la CLASIFICACIÓN DE RATING CON BASE EN TITLE

```python
# Build the RNN model
model_RNN_title = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=128, input_length=max_sequence_length),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(5, activation='softmax')  # Assuming ratings are from 1 to 5
])
model_RNN_title
```

Se compila y se entrena el modelo para la clasificación de Ratings con base en title 

```python
# Compile the model
model_RNN_title.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model_RNN_title.fit(X_title_padded_train, y_ratin_train - 1, epochs=10, batch_size=32, validation_split=0.2)  # Subtract 1 to convert 1-based ratings to 0-based index
```

Se evalua el modelo para la clasificación de Ratings con base en title 

```python
# Evaluate the model
loss_modelo_RNN_title, accuracy_modelo_RNN_title = model_RNN_title.evaluate(X_title_padded_test, y_ratin_test - 1)
print("Accuracy:", accuracy_modelo_RNN_title)
```

Ejemplo de predicicón con base en un título 

```python
X_title_padded_test[0]
```

```python
# Reshape the input to have a batch dimension
X_sample_title = X_title_padded_test[5]  # Replace with the appropriate sample
X_sample_title_reshape = X_sample_title.reshape(1, -1)  # Reshape to (1, sequence_length)

# Predict using the reshaped input
prediction_model_RNN_title = model_RNN_title.predict(X_sample_title_reshape)
prediction_model_RNN_title
```

```python
# Convert numerical sequence back to words
decoded_title_sample = tokenizer.sequences_to_texts([X_sample_title])[0]

print("Decoded title:", decoded_title_sample)
```

## Predicción del rating usando el review text

```python
# Split data into features (review_text) and target (ratings)
X_review = review_hoteles['review_text']
y_rating = review_hoteles['rating']
X_review
```

Llevamos todos los vectores de input al mismo tamanio que el vector de mayor tamanio

```python
max_sequence_length_review
```

```python
X_review_padded.shape
```

Vamos a separar en train y test

```python
from sklearn.model_selection import train_test_split

X_review_padded_train, X_review_padded_test, y_rating_train, y_rating_test = train_test_split(X_review_padded,y_rating,test_size=0.2,random_state=2023)
X_review_padded_train
```

```python
X_review_padded_train.shape
```

Construccion de la RNN para la CLASIFICACIÓN DE RATING CON BASE EN REVIEW_TEXT

```python
model_RNN_review2 = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=len(tokenizer_review.word_index) + 1, output_dim=264, input_length=max_sequence_length_review),
    tf.keras.layers.LSTM(128, return_sequences=True),  # Add return_sequences=True
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(5, activation='softmax')  # Assuming ratings are from 1 to 5
])
```

Se compila y se entrena el modelo para la clasificación de Ratings con base en review
