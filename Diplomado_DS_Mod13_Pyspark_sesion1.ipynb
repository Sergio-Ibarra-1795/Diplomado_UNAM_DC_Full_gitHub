```python
##Ver el contenido HDFS

! hdfs dfs -ls -R /tmp/dcd/ecobici/
```

```python
##Crear un DF con spark
eb_df = spark.read.csv('hdfs:///tmp/dcd/ecobici/', header=True, inferSchema=True)
```

```python
##Ver los 10 primeros renglones
eb_df.show(10)
```

```python
#Ver el numero de renglones, numero de columnas y las columnas del DF
print(f"Renglones {eb_df.count()}")
print(f"Columnas {len(eb_df.columns)}")
print(f"Columnas: {eb_df.columns}")
```

```python

##Crea un nuevo DF con solo 4 columnas
sub_eb_df = eb_df.select(['Genero_Usuario', 'Edad_Usuario', 'Bici', 'Fecha_Retiro'])
sub_eb_df.show(10)
print(f"Renglones {sub_eb_df.count()}")
```

```python
##Generar un nuevo DF quitando duplicados
sub_nd_eb_df = sub_eb_df.dropDuplicates()
print(f"Renglones {sub_nd_eb_df.count()}")
```

```python
##Ver estadisticos basicos
sub_nd_eb_df.describe().show()
```

```python
sub_nd_eb_df.write.save("gs://dcd05-sir-bucket/dcd/pyspark/ecobici/", format='csv', header=True)
```
